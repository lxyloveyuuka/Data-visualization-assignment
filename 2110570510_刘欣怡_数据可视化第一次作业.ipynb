```python
import networkx as nx
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = 'SimHei'

# 创建网络图对象
G = nx.Graph()

# 添加基础节点
G.add_nodes_from([
    (1, {"label": "数学系"}),
    (2, {"label": "计算机科学系"}),
    (3, {"label": "数据科学组"})
])

# 添加个性化节点
student_id = "2110570510"
student_name = "刘欣怡"
student_major = "大数据管理与应用"
interest_group_1 = "旅行"
interest_group_2 = "音乐"
interest_group_3 = "追星"
interest_group_2_1 = "吉他弹唱"
interest_group_2_2 = "R&B"
interest_group_1_1 = "南京"
interest_group_1_2 = "青岛"
interest_group_3_1 = "sugai yuuka"
interest_group_3_2 = "Taylor Swift"
interest_group_3_3 = "yurikanakamura"

G.add_nodes_from([
    (student_id, {"label": "2110570510"}),
    (student_name, {"label": "刘欣怡"}),
    (student_major, {"label": "大数据管理与应用"}),
    (interest_group_1, {"label": "旅行"}),
    (interest_group_2, {"label": "音乐"}),
    (interest_group_3, {"label": "追星"}),
    (interest_group_2_1,{"label": "吉他弹唱"}),
    (interest_group_2_2,{"label": "R&B"}),
    (interest_group_1_1, {"label": "南京"}),
    (interest_group_1_2, {"label": "青岛"}),
    (interest_group_3_1, {"label": "sugai yuuka"}),
    (interest_group_3_2, {"label": "Taylor Swift"}),
    (interest_group_3_3, {"label": "yurikanakamura"}),
])

# 添加基础边
G.add_edges_from([
    (1, 2),
    (2, 3)
])

# 添加个性化边
G.add_edges_from([
    (student_id, 2),
    (student_id, student_major),
    (student_id, student_name),
    (student_id, interest_group_1),
    (student_id, interest_group_2),
    (student_id, interest_group_3),
    (interest_group_2, interest_group_2_1),
    (interest_group_2, interest_group_2_2),
    (interest_group_1, interest_group_1_1),
    (interest_group_1, interest_group_1_2),
    (interest_group_3, interest_group_3_1),
    (interest_group_3, interest_group_3_2),
    (interest_group_3, interest_group_3_3)
])

# 定义节点类型对应的颜色
colors = {1: 'lightblue', 2: '#63B8FF', 3: 'lightblue', student_id: '#436EEE', student_major: '#63B8FF', 
          student_name: '#63B8FF', interest_group_1: '#63B8FF', interest_group_2: '#63B8FF', interest_group_3: '#63B8FF'}
default_color = 'lightblue'

# 为每个节点指定颜色
node_colors = [colors.get(node, default_color) for node in G.nodes()]

# 绘制网络图
pos = nx.spring_layout(G)
labels = {n: data['label'] for n, data in G.nodes(data=True)}
nx.draw_networkx_nodes(G, pos=pos, node_color=node_colors, node_size=450, alpha=0.6)#点的样式
nx.draw_networkx_edges(G, pos=pos, width=0.5, alpha=0.4)#边的样式
nx.draw_networkx_labels(G, pos=pos, labels=labels, font_size=10, font_color='#2B2B2B')#标签的样式
plt.axis('off')  # 关闭边框
plt.show()
```
